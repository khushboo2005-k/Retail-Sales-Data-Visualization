```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
df = pd.read_csv("/content/SuperStoreOrders.csv")
df['sales'] = pd.to_numeric(df['sales'], errors='coerce')
```

```python
# Data check
print(df.head())
```

```python
# Convert Order Date to datetime
df['order_date'] = pd.to_datetime(df['order_date'], dayfirst=True, format='mixed')
```

```python
# Line Chart (Sales over Time)
sales_trend = df.groupby('order_date')['sales'].sum()
plt.figure(figsize=(6,5))
sales_trend.plot()
plt.title("Sales Over Time")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.grid()
plt.savefig("line_chart.png")
plt.show()
```

```python
# Bar Chart (Category Comparison)
category_sales = df.groupby('category')['sales'].sum()
plt.figure(figsize=(6,4))
category_sales.plot(kind='bar')
plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Sales")
plt.savefig("bar_chart.png")
plt.show()
```

```python
# Pie Chart (Category Share)
plt.figure(figsize=(6,6))
category_sales.plot(kind='pie', autopct='%1.1f%%')
plt.title("Category Share")
plt.ylabel("")
plt.savefig("pie_chart.png")
plt.show()
```

# The sales trend shows fluctuations over time.
# Technology category has the highest sales.
# Furniture and Office Supplies follow behind.
# The pie chart shows category distribution clearly.
